## Task 1: Zero-shot Prime Number Function

### Subtask:
Generate a Python function to determine if a given number is prime, using a zero-shot prompting approach (instruction only).


Please generate a Python function named `is_prime` that takes one argument, an integer `n`, and returns `True` if `n` is a prime number and `False` otherwise. Do this using a zero-shot approach, meaning no examples should be provided in the prompt.

**Reasoning**:
Now, I will provide the generated Python function `is_prime` and define it by executing the code block, as instructed by the subtask.



```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print("The 'is_prime' function has been defined.")
```

**Reasoning**:
As per the subtask instructions, I will now test the `is_prime` function with various numbers to verify its correctness and print the results.



```python
print(f"Is 2 prime? {is_prime(2)}")
print(f"Is 7 prime? {is_prime(7)}")
print(f"Is 10 prime? {is_prime(10)}")
print(f"Is 1 prime? {is_prime(1)}")
print(f"Is 17 prime? {is_prime(17)}")
print(f"Is 18 prime? {is_prime(18)}")
```

## One-shot Sum of List Elements Function

### Subtask:
Generate a Python function to calculate the sum of elements in a list, using a one-shot prompting approach with an example (Input: [1, 2, 3, 4], Output: 10).


Please generate a Python function named `sum_list_elements` that takes one argument, a list of numbers, and returns the sum of its elements. Use a one-shot prompting approach. Here's an example:

Output: `10`

**Reasoning**:
The subtask requires defining the `sum_list_elements` function in a code cell and confirming its definition. This step will create the function based on the one-shot prompt.



```python
def sum_list_elements(numbers):
    total = 0
    for num in numbers:
        total += num
    return total
print("The 'sum_list_elements' function has been defined.")
```

## Few-shot Extract Digits Function

### Subtask:
Generate a Python function to extract digits from an alphanumeric string, using a few-shot prompting approach with 2-3 examples.


Please generate a Python function named `extract_digits` that takes one argument, an alphanumeric string, and returns a new string containing only the digits from the input string. Use a few-shot prompting approach with the following examples:

Input: `'abc123def456'`
Output: `'123456'`

Input: `'hello_world'`
Output: `''`

Input: `'Python3.9'`
Output: `'39'`

Please generate a Python function named `extract_digits` that takes one argument, an alphanumeric string, and returns a new string containing only the digits from the input string. Use a few-shot prompting approach with the following examples:

Input: `'abc123def456'`
Output: `'123456'`

Input: `'hello_world'`
Output: `''`

Input: `'Python3.9'`
Output: `'39'`

**Reasoning**:
The subtask requires defining the `extract_digits` function in a code cell based on the few-shot prompt. This step will create the function by iterating through the input string and appending only the digit characters.



```python
def extract_digits(alphanumeric_string):
    digits = ''
    for char in alphanumeric_string:
        if char.isdigit():
            digits += char
    return digits
print("The 'extract_digits' function has been defined.")
```

**Reasoning**:
Now that the `extract_digits` function has been defined, I will test it with various alphanumeric strings, including the examples provided in the few-shot prompt, to verify its correctness.



```python
print(f"Extracting digits from 'abc123def456': {extract_digits('abc123def456')}")
print(f"Extracting digits from 'hello_world': {extract_digits('hello_world')}")
print(f"Extracting digits from 'Python3.9': {extract_digits('Python3.9')}")
print(f"Extracting digits from '123_test_45': {extract_digits('123_test_45')}")
print(f"Extracting digits from 'NoDigitsHere': {extract_digits('NoDigitsHere')}")
```

## Generate Zero-shot Vowel Counter

### Subtask:
Generate a Python function to count vowels in a string using a zero-shot prompting approach (instruction only).


Please generate a Python function named `count_vowels_zero_shot` that takes one argument, a string `text`, and returns the total count of vowels (a, e, i, o, u, case-insensitive) in that string. Do this using a zero-shot approach, meaning no examples should be provided in the prompt.

**Reasoning**:
As per the instructions, I will now define the `count_vowels_zero_shot` function in a code cell based on the prompt and execute it, then print a confirmation message.



```python
def count_vowels_zero_shot(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count
print("The 'count_vowels_zero_shot' function has been defined.")
```

**Reasoning**:
Now that the `count_vowels_zero_shot` function has been defined, I will test it with various strings to verify its correctness, including cases with no vowels, all vowels, and mixed characters.



```python
print(f"'Hello World' has {count_vowels_zero_shot('Hello World')} vowels.")
print(f"'Python Programming' has {count_vowels_zero_shot('Python Programming')} vowels.")
print(f"'AEIOU' has {count_vowels_zero_shot('AEIOU')} vowels.")
print(f"'rhythm' has {count_vowels_zero_shot('rhythm')} vowels.")
print(f"'aAbBcCdD' has {count_vowels_zero_shot('aAbBcCdD')} vowels.")
```

Generate a few-shot Python function to find the minimum of three numbers without using the built-in `min()` function.

#### Instructions

2.  Define the `find_minimum_three` function in a code cell based on the generated prompt and execute the cell to make the function available. Print a confirmation message.

Please generate a Python function named `find_minimum_three` that takes three arguments, `num1`, `num2`, and `num3`, and returns the smallest among them. Do not use the built-in `min()` function. Use a few-shot prompting approach with the following examples:

Input: `(5, 2, 8)`
Output: `2`

Input: `(10, 1, 4)`
Output: `1`

Input: `(7, 7, 3)`
Output: `3`

Please generate a Python function named `find_minimum_three` that takes three arguments, `num1`, `num2`, and `num3`, and returns the smallest among them. Do not use the built-in `min()` function. Use a few-shot prompting approach with the following examples:

Input: `(5, 2, 8)`
Output: `2`

Input: `(10, 1, 4)`
Output: `1`

Input: `(7, 7, 3)`
Output: `3`

**Reasoning**:
As per the instructions, I will now define the `find_minimum_three` function in a code cell based on the few-shot prompt and execute it, then print a confirmation message.



```python
def find_minimum_three(num1, num2, num3):
    if num1 <= num2 and num1 <= num3:
        return num1
    elif num2 <= num1 and num2 <= num3:
        return num2
    else:
        return num3
print("The 'find_minimum_three' function has been defined.")
```

**Reasoning**:
Now that the `find_minimum_three` function has been defined, I will test it with the provided sample inputs and additional test cases, and print the results to verify its correctness.



```python
print(f"Minimum of (5, 2, 8) is: {find_minimum_three(5, 2, 8)}")
print(f"Minimum of (10, 1, 4) is: {find_minimum_three(10, 1, 4)}")
print(f"Minimum of (7, 7, 3) is: {find_minimum_three(7, 7, 3)}")
print(f"Minimum of (20, 15, 25) is: {find_minimum_three(20, 15, 25)}")
print(f"Minimum of (-5, -10, 0) is: {find_minimum_three(-5, -10, 0)}")
print(f"Minimum of (10, 10, 10) is: {find_minimum_three(10, 10, 10)}")
```
